# **Simple Moving Average (SMA)**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Assuming 'df' is your original DataFrame with 'Date' as index and 'AveragePrice' as column
# Reset the index to get 'Date' back as a column
df = df.reset_index()

# Convert 'Date' column to datetime objects
df['Date'] = pd.to_datetime(df['Date'])

# Set 'Date' as index again
df = df.set_index('Date')

# Use data from 2015 to 2016 for training
# Filter by year instead of using loc with date strings
train_data = df[(df.index >= pd.to_datetime('2015-01-01')) & (df.index <= pd.to_datetime('2016-12-31'))]

# Resample to monthly frequency and calculate mean price for training data
monthly_train_data = train_data['AveragePrice'].resample('M').mean()

# Forecast using a simple moving average (SMA) approach (Non-Regression Model)
def moving_average_forecast(data, window, horizon):
    forecast = []
    for i in range(horizon):
        avg = np.mean(data[-window:])  # Take average of last 'window' months
        forecast.append(avg)
        data = np.append(data, avg)  # Append forecast to data for next iteration
    return forecast

# Set window size and forecast horizon
window_size = 3  # Using the last 3 months for moving average
forecast_horizon = 12  # Forecast for 12 months of 2017
forecast = moving_average_forecast(monthly_train_data.values, window_size, forecast_horizon)

# Create a date range for the forecast (2017)
forecast_dates = pd.date_range(start='2017-01-01', periods=forecast_horizon, freq='MS')

# Get actual data for 2017
# Filter by year instead of using loc with date strings
actual_data_2017 = df[(df.index >= pd.to_datetime('2017-01-01')) & (df.index <= pd.to_datetime('2017-12-31'))]['AveragePrice'].resample('M').mean()

# Calculate average price for each month across all years
average_prices_by_month = df['AveragePrice'].groupby(df.index.month).mean()

# Create the plot
plt.figure(figsize=(12, 6))

# Plot average price for each month
plt.plot(forecast_dates, average_prices_by_month[forecast_dates.month],
         label='Actual Prices ', marker='o')

# Commented out the actual price line for 2017
# plt.plot(actual_data_2017.index, actual_data_2017.values, label='Actual Prices (2017)', marker='o')

# Plot predicted data for 2017
plt.plot(forecast_dates, forecast, label='Predicted Prices', marker='o', linestyle='--')

# Formatting and labels
plt.xlabel('Date')
plt.ylabel('Average Price')
plt.title('Avocado Price Forecast for 2017 (Trained on 2015-2016)')
plt.grid(True)
plt.legend()

# Set x-axis ticks to show only month names (January to December)
plt.xticks(forecast_dates, forecast_dates.strftime('%B'), rotation=45, ha='right')

plt.tight_layout()
plt.show()
```
